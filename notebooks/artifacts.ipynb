```python
import pandas as pd
import numpy as np
from pathlib import Path

ARTIFACTS_DIR = Path.home() / "icidea_llm_ids" / "artifacts"
RESULTS_DIR   = Path.home() / "icidea_llm_ids" / "results"

print("LOCAL ARTIFACTS INVENTORY")
print("="*60)

# Expected files
expected = {
    # HCRL Qwen P1-P5
    "section9a_classifier_results.parquet":     "SecureBERT results",
    "section9b_baseline_explanations.parquet":  "Qwen baseline",
    "section10_perturbed_p1p2.parquet":         "Qwen HCRL P1+P2",
    "section10_perturbed_p3p4p5.parquet":       "Qwen HCRL P3+P4+P5",
    "section10_p6_false_context.parquet":       "P6 dataset",
    "section11_metrics.parquet":                "Qwen HCRL metrics",

    # HCRL Mistral P6
    "mistral_hcrl_p6_full.parquet":             "Mistral HCRL P6",

    # ROAD Qwen
    "task4_road_windows.parquet":               "ROAD windows",
    "task4_road_results.parquet":               "Qwen ROAD P1",
    "task4_road_p5_results.parquet":            "Qwen ROAD P5",
    "task4_road_p6_dataset.parquet":            "ROAD P6 dataset",

    # Mistral ROAD
    "mistral_road_p5_full.parquet":             "Mistral ROAD P5",
    "mistral_road_p6_false_context_full.parquet": "Mistral ROAD P6",

    # Validation
    "task1_shap_comparison.parquet":            "SHAP dissociation",
    "task2_seed_results.parquet":               "Seed stability",
    "p6_dissociation_verification.parquet":     "P6 dissociation",

    # Qwen HCRL P6
    "Qwen HCRL Full.parquet":                   "Qwen HCRL P6",
}

found    = []
missing  = []

for fname, description in expected.items():
    path = ARTIFACTS_DIR / fname
    if path.exists():
        size_mb = path.stat().st_size / 1e6
        try:
            df = pd.read_parquet(path)
            found.append((fname, description,
                         size_mb, len(df)))
            print(f"  ✓ {description:<35} "
                  f"{len(df):>6} rows  "
                  f"{size_mb:.2f} MB")
        except Exception as e:
            print(f"  ⚠ {description:<35} "
                  f"EXISTS but unreadable: {e}")
    else:
        missing.append((fname, description))
        print(f"  ✗ {description:<35} MISSING")

print(f"\n{'='*60}")
print(f"Found:   {len(found)}/{len(expected)}")
print(f"Missing: {len(missing)}/{len(expected)}")

if missing:
    print(f"\nMissing files:")
    for fname, desc in missing:
        print(f"  {desc}: {fname}")

# Check results directory
print(f"\n\nRESULTS DIRECTORY")
print("="*60)
for f in sorted(RESULTS_DIR.glob("*.parquet")):
    size_mb = f.stat().st_size / 1e6
    try:
        df = pd.read_parquet(f)
        print(f"  ✓ {f.name:<45} "
              f"{len(df):>6} rows  "
              f"{size_mb:.2f} MB")
    except:
        print(f"  ⚠ {f.name} — unreadable")

# Summary of total data
print(f"\n\nDATA SUMMARY")
print("="*60)
total_rows = sum(rows for _, _, _, rows in found)
total_mb   = sum(mb for _, _, mb, _ in found)
print(f"Total records across all files: {total_rows:,}")
print(f"Total size: {total_mb:.1f} MB")
```
